```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def Parsing_url(url_link):
    # Making a GET request
    r=requests.get(url_link)

    #Parsing the HTML
    soup=BeautifulSoup(r.content,'html.parser')
    return soup

def inside_parsing_url(sub_url):
    r=requests.get(sub_url)

    #Parsing the HTML
    soup_1=BeautifulSoup(r.content,'html.parser')
    return soup_1
```
