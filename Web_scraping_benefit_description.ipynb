```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def Parsing_url(url_link):
    # Making a GET request
    r=requests.get(url_link)

    #Parsing the HTML
    soup=BeautifulSoup(r.content,'html.parser')
    return soup

def inside_parsing_url(sub_url):
    r=requests.get(sub_url)

    #Parsing the HTML
    soup_1=BeautifulSoup(r.content,'html.parser')
    return soup_1
```

```python
#u="https://www.insuranceopedia.com/definition/920/back-load"

url='https://www.insuranceopedia.com/dictionary/'
soup=Parsing_url(url)
```

```python
container=soup.find("section", id="dictionaryTerms")
container
```

```python
final_df_1=pd.DataFrame()
terms = container.find_all('a')
for term in terms:
    term_text = term.text.strip()
    term_link = term['href']
    inside_link="https://www.insuranceopedia.com"+term_link
    try:
      html_doc=Parsing_url(inside_link)
      internal_term = html_doc.find_all("div", class_="jan-article__content")
      print("Term:", term_text)
      print("Link:", term_link)
      #print(sub_main_class)
      #combined_description=" "
      for i in internal_term:
          description = i.find("p").text.strip()
          out_df=pd.DataFrame({"Benefit_name":term_text,
                                  "Web_link":inside_link,
                                  "Description":description},index=[0])
          final_df_1=final_df_1.append(out_df)
    except Exception as e:
            continue
final_df_1.reset_index(inplace=True,drop=True)
#         if description:
#             combined_description+=description.text.strip()
        
        #if description!= None:
        #combined_description += description
    #if len(combined_description)==2:

        #print("Description :", description)
        
#         if len(combined_description) >= 2:
#             description = combined_description[1:]#.get_text(separator=' ')
#             print("best_description :", description)
#         print()
```
