We have created Web_scraping for this website:  url='https://www.healthcare.gov/glossary/'

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def Parsing_url(url):
    # Making a GET request
    r=requests.get(url)

    #Parsing the HTML
    soup=BeautifulSoup(r.content,'html.parser')
    return soup

def inside_parsing_url(sub_url):
    r=requests.get(sub_url)

    #Parsing the HTML
    soup_1=BeautifulSoup(r.content,'html.parser')
    return soup_1
```

```python
url='https://www.healthcare.gov/glossary/'
soup=Parsing_url(url)
```

```python
main= soup.find('main', id ='main')
containers=main.find_all('div', class_='ds-l-container')[2]
print(containers)
```

```python
final_df=pd.DataFrame()
for container in containers:
    #container=list(container)
#y=container.find_all(class_="termsContainer-module--terms-container--f8c7a")
    term_list = container.find('ul', class_='termsContainer-module--term-list--bb3a8')
    terms = term_list.find_all('a', class_='termsContainer-module--term-link--6da2c')
    for term in terms:
        term_text = term.text.strip()
        term_link = term['href']
        a="https://www.healthcare.gov"
        inside_link=a+term_link
        #print(inside_link)
        print("inside_link_1", inside_link)
        try:
            sub_soup=Parsing_url(inside_link)
            #print(sub_soup)
            inside_find = sub_soup.find('main', id ='main')
            #print(inside_find)
            description = inside_find.find("div", class_="index-module--content-container--60994").text.strip()
            out_df=pd.DataFrame({"Benefit_name":term_text,
                                 "Web_link":inside_link,
                                 "Description":description},index=[0])
            final_df=final_df.append(out_df)
    #final_df.reset_index(inplace=True,drop=True)
            
        #final_df.reset_index(inplace=True,drop=True)
        except Exception as e:
            continue
        #final_df=final_df.append(out_df)
    #final_df.set_index(inplace=True)#,drop=True)
final_df.reset_index(inplace=True,drop=True)
#         print("Term:", term_text)
#         print("Link:", term_link)
#         print("inside_link", inside_link)
#         print("\n")
#         print("Description:", description)
#         print()
        #final_df.reset_index(inplace=True,drop=True)
        
        #print("sub_soup", sub_soup)
```
